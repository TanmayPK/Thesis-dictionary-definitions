```python
file_path = 'Data/definitions_dataset/train.json'
with open(file_path, 'r', encoding='utf-8') as file:
    json_data_test = json.load(file)

flattened_data = [{'word': sublist[0][0], 'definition': ' '.join(sublist[1]), 'context': ' '.join(sublist[2])}
                  for sublist in json_data_test]
eng = pd.DataFrame(flattened_data)
```

```python
eng.head()
```

```python
lemmatizer = WordNetLemmatizer()

def word_in_context(word, context):
    context_lower = context.lower()
    if word in context_lower.split():
        return True
    root_word = lemmatizer.lemmatize(word)
    if root_word in context_lower.split():
        return True
    return False

filtered_eng_test = eng_test[eng_test.apply(lambda row: word_in_context(row['word'], row['context']), axis=1)]
```

```python
filtered_eng_test.reset_index(drop=True, inplace=True)
```

```python
def extract_first_tekst(xml):
    soup = BeautifulSoup(xml, 'xml')
    voorbeeld_tags = soup.find_all(['definitie', 'Definitie'])
    for tag in voorbeeld_tags:
        return tag.text.strip()
    return None
```

```python
eng_prompt = """example: {}

word: {}

provide a dictionary definition for the word based on the example.

definition: """
```

```python
def apply_prompt(df, context_col, word_col):
    df['prompt'] = df.apply(lambda row: eng_prompt.format(row[context_col], row[word_col]), axis=1)
    return df
```

```python
eng_train = apply_prompt(filtered_eng_train, 'context', 'word')
eng_test = apply_prompt(filtered_eng_test, 'context', 'word')
eng_val = apply_prompt(filtered_eng_val, 'context', 'word')
```
